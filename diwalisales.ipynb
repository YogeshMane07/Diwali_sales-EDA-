```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('Diwali Sales Data.csv',encoding="ISO-8859-1")
df.shape
df.info()
df.head()
df.columns
df.describe()
df.drop(columns=['Status','unnamed1'],inplace=True)
df.columns
pd.isnull(df)
df.isnull().sum()
df['Amount'].fillna(df['Amount'].median(), inplace=True)
df.isnull().sum()

df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.shape
df.describe()
df.describe(include=['object'])
plt.figure(figsize=(9,3))
plt.scatter(df['Gender'],df['Orders'])
plt.xlabel('Gender')
plt.ylabel('Orders')
plt.show()
plt.figure(figsize=(9,3))
plt.scatter(df['Age'],df['Orders'])
plt.xlabel('Age')
plt.ylabel('Orders')
plt.show()
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="Age Group", y="Amount", hue="Gender")
plt.title("Scatter Plot")
plt.xlabel("Age Group")
plt.ylabel("Amount Spent")
plt.legend(title="Gender")
plt.show()
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x="Gender")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x="Age Group",order=df['Age Group'].value_counts().index)
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(10, 8))
sns.countplot(data=df, y="Product_Category", order=df["Product_Category"].value_counts().index)
plt.title("Product Category")
plt.xlabel("Count")
plt.ylabel("Product Category")
plt.show()
plt.figure(figsize=(14, 6))
state_sales = df.groupby("State")["Amount"].sum().sort_values(ascending=False)
state_sales.plot(kind="bar")
plt.title("Total Sales by State")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=50)
plt.show()
plt.figure(figsize=(8, 8))
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%0.1f%%')
plt.title("Gender Distribution of Customers")
plt.show()
plt.figure(figsize=(8, 8))
occ_counts = df['Occupation'].value_counts()
plt.pie(occ_counts, labels=occ_counts.index, autopct='%0.1f%%')
plt.title("Occupation Distribution of Customers")
plt.show()
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x="Age Group",order=df['Age Group'].value_counts().index,hue='Gender')
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Gender", y="Amount")
plt.title("Box Plot of Gender")
plt.xlabel("Gender")
plt.ylabel("Amount")
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Age Group", y="Amount")
plt.title("Box Plot of Age Group")
plt.xlabel("Age Group")
plt.ylabel("Amount")
plt.show()

plt.figure(figsize=(20,9))
sns.countplot(data = df, x = 'Occupation')
plt.title('Occupation Chart')
plt.xlabel('Occupation')
state_sales = df.groupby(['Marital_Status', 'Gender'], as_index = False).sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(data = state_sales, x = 'Marital_Status', y = 'Amount', hue='Gender')
plt.figure(figsize=(20,9))
sns.countplot(data = df, x = 'Marital_Status')
plt.xlabel('Marital_Status')
plt.ylabel('count')
df.columns
df.drop(labels=['Cust_name', 'Product_ID', 'Gender', 'Age Group','State', 'Zone', 'Occupation', 'Product_Category'],axis=1,inplace=True)
df.columns
df.head
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, fmt=".2f")
plt.title("Heatmap")
plt.show()
```
